```python
import numpy as np

#Features:[Study hours,Attendence]
X=np.array([[2,40],[3,45],[4,50],[5,55],[6,60],[7,75],[8,78],[9,90],[10,100]])

#Labels:0-fail,1-pass
Y=np.array([0,0,0,1,1,1,1,1,1])
```

# **Build SVM model**

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

#split the data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)

#Create svm model
model=SVC(kernel='linear')

#train model
model.fit(X_train,Y_train)
```

# **Test in real time**

```python
new_student=[[11,150]]
prediction=model.predict(new_student)
if prediction[0]==1:
  print("Student will PASS")
else:
  print("Student will FAIL")
```

# **To visual **

```python
import matplotlib.pyplot as plt
fail=X[Y==0]
pass_=X[Y==1]

plt.scatter(fail[:,0],fail[:,0],label='fail',marker='x')
plt.scatter(pass_[:,1],pass_[:,1],label='pass',marker='o')

#plot the new student
plt.scatter(11,150,label='new student',marker='*',s=200)
plt.xlabel("Study Hours")
plt.ylabel("Attendence")
plt.title("Student performance classification using svm")
plt.legend()
plt.grid(True)
plt.show()
```
